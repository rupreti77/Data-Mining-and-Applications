**CLEANING THE DATA**

```python
import re
import string

#This function converts to lower-case, removes square bracket, removes numbers and punctuation
def text_clean_1(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

cleaned1 = lambda x: text_clean_1(x)
```

```python
opinions4 = opinions3[['userID','commentBody']].copy()
```

```python
opinions4['cleanedcomment'] = pd.DataFrame(opinions4.commentBody.apply(cleaned1))
opinions4.head(100)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
vds = SentimentIntensityAnalyzer()
```

```python
opinions4['compound'] = [vds.polarity_scores(x)['compound'] for x in opinions4['cleanedcomment']]
opinions4['neg'] = [vds.polarity_scores(x)['neg'] for x in opinions4['cleanedcomment']]
opinions4['neu'] = [vds.polarity_scores(x)['neu'] for x in opinions4['cleanedcomment']]
opinions4['pos'] = [vds.polarity_scores(x)['pos'] for x in opinions4['cleanedcomment']]
```

```python
opinions4
```

```python
opinions4['sentiment'] = ['Positive' if x >= 0.05 else 'Negative' if x <=-0.05 else 'Neutral' for x in opinions4['compound']]
opinions4
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
print('Percentage for default\n')
print(round(opinions4.sentiment.value_counts(normalize=True)*100,2))
round(opinions4.sentiment.value_counts(normalize=True)*100,2).plot(kind='bar')
plt.title('Percentage Distributions by review type')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

Independent_var = opinions4.cleanedcomment
Dependent_var = opinions4.sentiment

IV_train, IV_test, DV_train, DV_test = train_test_split(Independent_var, Dependent_var, test_size = 0.2, random_state = 225)

print('IV_train :', len(IV_train))
print('IV_test  :', len(IV_test))
print('DV_train :', len(DV_train))
print('DV_test  :', len(DV_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

tvec = TfidfVectorizer()
clf2 = LogisticRegression(solver = "lbfgs")


from sklearn.pipeline import Pipeline
```

```python
model = Pipeline([('vectorizer',tvec),('classifier',clf2)])

model.fit(IV_train, DV_train)


from sklearn.metrics import confusion_matrix

predictions = model.predict(IV_test)

confusion_matrix(predictions, DV_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print("Accuracy : ", accuracy_score(predictions, DV_test))
print("Precision : ", precision_score(predictions, DV_test, average = 'weighted'))
print("Recall : ", recall_score(predictions, DV_test, average = 'weighted'))
```
